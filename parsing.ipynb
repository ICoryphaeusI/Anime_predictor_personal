```python

from time import sleep
import re
import openpyxl
import requests
from bs4 import BeautifulSoup

# Загрузка существующего Excel-файла
existing_workbook = openpyxl.load_workbook('anime_links.xlsx')
existing_sheet = existing_workbook.active

# Получение значений из второго столбца (второй столбец имеет индекс 1)
urls_column = existing_sheet['B']  # Здесь B - это буква столбца

main_data = []

for index, url_cell in enumerate(urls_column, start=1):  # Начинаем с индекса 1, так как первая ячейка заголовка пропущена
    url = url_cell.value  # Получаем значение URL из ячейки
    sleep(1)
    print(f"Элемент {index}: {url}")
    
    try:#Находит название тайтла
        second_response = requests.get(url, headers=headers, timeout=5)
        second_soup = BeautifulSoup(second_response.text, 'lxml')
        anime_name = second_soup.find('header', class_='head').text.split("/")[0][:-1]
    except:
        anime_name = 'Сайт умер...'
    try:#Создаёт список со всеми характеристиками левого блока
        char_list = second_soup.select('.c-info-left .block .line-container')
    except:
        print(f'Не смог взять характеристики для {url}')
    try:#Находит тип аниме
        anime_type = next((i.text[7:-2] for i in char_list if 'Тип' in str(i)), 'Тип не указан')
    except:
        anime_type = "Не смог"
        print(f'Не смог найти тип для аниме {url}')
    try:#Находит кол-во эпизодов
        episodes = next((i.text.split(':')[1][1:-2] for i in char_list if "Эпизоды" in str(i)), 'Не указано')
    except:
        episodes = 'Не смог'
        print(f'Не смог найти кол-во эпизодов для аниме {url}')
    try:#Находит длину эпизода
        episode_length = next((i.text.split(':')[1][1:-3] for i in char_list if 'Длительность эпизода:' in str(i)), 'Не указано')
    except:
        episode_length = 'Не смог'
        print(f'Не смог найти длину эпизода(ов) для аниме {url}')
    try:#Находит жанры тайтла
        genres = next((re.findall('[А-ЯЁ][а-яё]*', i.text)[1:] for i in char_list if "Жанр" in str(i)), 'Не указано')
    except:
        genres = 'Не смог'
        print(f'Не смог найти жанры для аниме {url}')
    try:#Находит возрастное ограничение
        age_limit = next((i.text.split(':')[1][1:-2] for i in char_list if 'Рейтинг' in str(i)), 'Не указано')
    except:
        age_limit = 'Не смог'
        print(f'Не смог найти возрастное ограничение для аниме {url}')

    try:#Находит студию
        studio = second_soup.find('div', class_='c-info-right')
        string = str(studio)
        start = string.index('Аниме студии')
        result = string[start:]
        new_result = result.split('">')[0]
        studio = new_result[13:]
    except:
        studio = 'Не смог'
        print(f'Не смог найти студию для аниме {url}')
    
    try:#Находит автора
        res_link = url + '/resources'
        response_third = requests.get(res_link, headers=headers, timeout=5)
        soupp = BeautifulSoup(response_third.text, 'lxml')
        author = soupp.select_one('.c-authors .cc .info .name-en').text
        
    except:
        author = 'Не смог'
        print(f'Не смог найти автора для аниме {url}')

    if author == 'Не смог':
        try:
            res_link = url + '/resources'
            response_third = requests.get(res_link, headers=headers, timeout=5)
            soupp = BeautifulSoup(response_third.text, 'lxml')
            auth = soupp.find('div', class_='c-column c-authors block_m').find_all('div', class_='b-db_entry-variant-list_item')
            string = str(auth[0])
            string = string.split('data-url')[1]
            string = string.split('><div class')[0]
            string = string[2:-1]
            response = requests.get(string, headers=headers)
            soup = BeautifulSoup(response.text, 'lxml')
            author = soup.find('header', class_='head').find('meta').get('content')
            print(f'Смог найти автора для аниме {url}')
        except:
            author = 'Не смог X2'
            print(f'Не смог найти автора X2 для аниме {url}')

    try:#Находит рейтинг татйла
        score = second_soup.find('div', class_='text-score').text
        digits_only = re.sub(r'\D', '', score)
        rate = digits_only[:1] + '.' + digits_only[1:]
    except:
        rate = 'Не смог'
        print(f'Не смог найти рейтинг для аниме {url}')

    main_data.append([anime_name, anime_type, episodes, episode_length, genres, age_limit, studio, author, rate, url])


```

```python
import openpyxl

# Загружаем файл
workbook = openpyxl.load_workbook('shiki13.xlsx')
sheet = workbook.active

# Создаем словарь, где ключами будут элементы первого столбца, а значениями - строки
data_dict = {}
for row in sheet.iter_rows(min_row=2, values_only=True):  # Пропускаем заголовок
    title = str(row[0]).strip() if row[0] else ""
    data_dict.setdefault(title, []).append('\t'.join(str(cell) if cell is not None else "" for cell in row))

# Выводим строки с одинаковыми элементами в первом столбце
for title, rows in data_dict.items():
    if len(rows) > 1:
        print(f"Заголовок: {title}")
        for row in rows:
            print(row)
        print('=' * 50)
```

```python
workbook = openpyxl.load_workbook('shiki13.xlsx')
sheet = workbook.active

# Создаем словарь для подсчета вхождений строк
row_counts = {}

# Подсчитываем вхождения строк
for row in sheet.iter_rows(min_row=2, values_only=True):  # Пропускаем заголовок
    row_data = tuple(str(cell) if cell is not None else "" for cell in row)
    row_counts[row_data] = row_counts.get(row_data, 0) + 1

# Выводим строки и количество вхождений
for row_data, count in row_counts.items():
    if count > 1:
        formatted_row = '\t'.join(row_data)
        print("Строка: {}, Количество вхождений: {}".format(formatted_row, count))
```
